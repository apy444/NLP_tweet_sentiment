```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sentiment-model/test_sentiment.csv')
```

```python
df
```

```python
df['date'] = df['year'].astype(str) + '-' + df['month'].map('{:02d}'.format).astype(str) + '-' + df['day'].map('{:02d}'.format).astype(str)
df['positive_tweets'] = np.where(df['sentiment'] == 1, 1, 0)
df['negative_tweets'] = np.where(df['sentiment'] != 1, 1, 0)
```

```python
df
```

```python
agg = df[['ticker_symbol', 'company_name', 'sector', 'date', 'positive_tweets', 'negative_tweets']].groupby(['ticker_symbol', 'company_name', 'sector', 'date']).sum().reset_index()
```

```python
agg = agg.sort_values(by=['company_name', 'date'])
agg
```

```python
agg.to_csv('ui_sentiment.csv', index=False)
```
